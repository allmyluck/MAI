```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
import math
warnings.simplefilter("ignore")
%matplotlib inline

sns.set(style="darkgrid")
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split

data = data.drop(columns=["Id"])

y = data["SalePrice"]
X = data.drop(columns=["SalePrice"])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
sns.distplot(y_train)
```

```python
# Обучите линейную регрессию (LinearRegression, Ridge, Lasso) на числовых признаках из нашего датасета
# TO DO
from sklearn.linear_model import LinearRegression, Ridge, Lasso

numeric_features = list(X_train.dtypes[X_train.dtypes == "float64"].index) + list(X_train.dtypes[X_train.dtypes == "int64"].index)

X_train = X_train.fillna(0)
X_test = X_test.fillna(0)

lregr = LinearRegression().fit(X_train[numeric_features], y_train)
r_regr = Ridge().fit(X_train[numeric_features], y_train)
las_regr = Lasso().fit(X_train[numeric_features], y_train)
```

$$\text{RMSE}(X, y, a) = \sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} (y_i - a(x_i))^2}$$

```python
# Определите RMSE через функцию sklearn.metrics.make_scorer
# TO DO
from sklearn.metrics import make_scorer

def rmse(true, predictions):
    return math.sqrt(np.square(np.subtract(true,predictions)).mean())
loss_rmse  = make_scorer(rmse, greater_is_better=False)
score_rmse = make_scorer(rmse, greater_is_better=True)
```

```python
def show_weights(features, weights, scales):
    fig, axs = plt.subplots(figsize=(14, 10), ncols=2)
    sorted_weights = sorted(zip(weights, features, scales), reverse=True)
    weights = [x[0] for x in sorted_weights]
    features = [x[1] for x in sorted_weights]
    scales = [x[2] for x in sorted_weights]
    sns.barplot(y=features, x=weights, ax=axs[0])
    axs[0].set_xlabel("Weight")
    sns.barplot(y=features, x=scales, ax=axs[1])
    axs[1].set_xlabel("Scale")
    plt.tight_layout()
```

```python
#show_weights(numeric_features, model.coef_, X_train[numeric_features].std())
show_weights(numeric_features, lregr.coef_, X_train[numeric_features].std())
show_weights(numeric_features, r_regr.coef_, X_train[numeric_features].std())
show_weights(numeric_features, las_regr.coef_, X_train[numeric_features].std())
```

```python
# Отмасштабируйте признаки перед обучением модели, воспользуясь трансформером StandardScaler
# TO DO

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train[numeric_features])
X_train[numeric_features] = scaler.transform(X_train[numeric_features])
```

```python
#scales = pd.Series(data=X_train_scaled.std(axis=0), index=numeric_features)
#show_weights(numeric_features, model.coef_, scales)
scales = pd.Series(data=X_train[numeric_features].std(axis=0), index=numeric_features)
show_weights(numeric_features, lregr.coef_, X_train[numeric_features].std())
show_weights(numeric_features, r_regr.coef_, X_train[numeric_features].std())
show_weights(numeric_features, las_regr.coef_, X_train[numeric_features].std())
```

```python
# Подберите коэффициент регуляризации alpha, воспользуясь GridSearchCV
# TO DO

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

gbr_grid_search = GridSearchCV(cv = 5, estimator= Lasso(), 
                               param_grid = [{'alpha': [0, 0.0001, 0.001, 0.1, 1, 10,100]}],
                               scoring = make_scorer(r2_score)
                              )

gbr_grid_search.fit(X_train[numeric_features], y_train)

print(gbr_grid_search.best_params_)
print(gbr_grid_search.best_score_)
print(gbr_grid_search.best_estimator_)

gbr_grid_search = GridSearchCV(cv = 5, estimator= Ridge(), 
                               param_grid = [{'alpha': [0, 0.0001, 0.001, 0.1, 1, 10,100]}],
                               scoring = make_scorer(r2_score)
                              )
gbr_grid_search.fit(X_train[numeric_features], y_train)

print(gbr_grid_search.best_params_)
print(gbr_grid_search.best_score_)
print(gbr_grid_search.best_estimator_)
```

```python
# Обучите линейную регрессию (LinearRegression, Ridge, Lasso) с подобранным коэффициентом регуляризации на числовых признаках из нашего датасета, воспользуясь классом Pipeline (не забывать про мастабирование признаков через StandardScaler)
# TO DO

from sklearn.pipeline import Pipeline

lasso_pipe = Pipeline([('scaler', StandardScaler()), ('regressor', Lasso(alpha= 1000))])
lasso_pipe.fit(X_train[numeric_features], y_train)
print(lasso_pipe.score(X_train[numeric_features], y_train))



pipe = Pipeline([('scaler', StandardScaler()), ('regressor', Ridge(alpha= 100))])
pipe.fit(X_train[numeric_features], y_train)
print(pipe.score(X_train[numeric_features], y_train))

```

```python
categorical = list(X_train.dtypes[X_train.dtypes == "object"].index)
X_train[categorical] = X_train[categorical].fillna("NotGiven")
X_test[categorical] = X_test[categorical].fillna("NotGiven")
```

```python
X_train[categorical].sample(5)
```

```python
# Закодируйте категориальные признаки числами с помощью OneHotEncoder и обучите модели на всех признаках
def get_one_hot(data):
    for column in categorical:
        tempdf = pd.get_dummies(data[column], prefix=column)
        data = pd.merge(
              left=data,
              right=tempdf,
              left_index=True,
              right_index=True,
        )
    data = data.drop(categorical, axis=1)
    return data

X = get_one_hot(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)
print("Size before OneHot:", X_train.shape)
X_train.head()
```

```python
# Обучите линейную регрессию с l1 и l2 регуляризаторами
lasso_pipe = Pipeline([('scaler', StandardScaler()), ('lasso', Lasso(alpha= 1000))])
lasso_pipe.fit(X_train, y_train)
print(lasso_pipe.score(X_train, y_train))

pipe = Pipeline([('scaler', StandardScaler()), ('ridge', Ridge(alpha= 100))])
pipe.fit(X_train, y_train)
print(pipe.score(X_train, y_train))
```

```python
ridge_zeros = np.sum(pipe.steps[-1][-1].coef_ == 0)
lasso_zeros = np.sum(lasso_pipe.steps[-1][-1].coef_ == 0)
print("Zero weights in Ridge:", ridge_zeros)
print("Zero weights in Lasso:", lasso_zeros)
```

```python
# Подберите коэффициент регуляризации для всех моделей, воспользуясь GridSearchCV
# TO DO
gbr_grid_search = GridSearchCV(cv = 5, estimator= Lasso(), 
                               param_grid = [{'alpha': [0, 0.0001, 0.001, 0.1, 1, 10,100, 1000, 10000]}],
                               scoring = make_scorer(r2_score)
                              )

gbr_grid_search.fit(X_train, y_train)

print(gbr_grid_search.best_params_)
print(gbr_grid_search.best_score_)
print(gbr_grid_search.best_estimator_)

gbr_grid_search = GridSearchCV(cv = 5, estimator= Ridge(), 
                               param_grid = [{'alpha': [0, 0.0001, 0.001, 0.1, 1, 10,100, 1000, 10000]}],
                               scoring = make_scorer(r2_score)
                              )
gbr_grid_search.fit(X_train, y_train)

print(gbr_grid_search.best_params_)
print(gbr_grid_search.best_score_)
print(gbr_grid_search.best_estimator_)
```

```python
pipe = Pipeline([('scaler', StandardScaler()), ('ridge', Ridge(alpha= 10))])
pipe.fit(X_train, y_train)
pipe.score(X_train, y_train)
error = (y_train - pipe.predict(X_train)) ** 2
sns.distplot(error)
```
